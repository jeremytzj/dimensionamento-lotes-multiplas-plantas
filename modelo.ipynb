O nosso objetivo é maximizar o atendimento com o menor custo de frete possível e com o menor volume “parado” em estoque.

#Conjuntos
$⊂ ℕ$

$P = \{1,2,3\}:$ conjunto de plantas

$L_p = \{1_0,2_0,1_1,…,4_1,1_2, …,3_2 \}:$ conjunto de linhas

$W = \{ 1,\dots , 5 \}:$ conjunto de armazéns externos

$C = \{0,1,…,14\}:$ conjunto de clientes

$I = \{ 1,2,…, 25 \}:$ conjunto de produtos

$E_I = \{1,2,…,5 \}:$ tamanho do produto $i$

$T = \{0, 1, \dots , 6 \}:$ conjunto de períodos

# Variáveis de decisão

$ \in ℕ $

$x_{i_e,l_p,t}:$ quantidade de produto $i_e$ (tamanho e) produzido na linha $l_p$ (da planta p) no período t

$S^{P}_{i,p,t}:$ quantidade de produto i estocado na planta p no período t

$S^{W}_{i,w,t}:$ quantidade de produto i estocado no armazém w no período t

$f^{PC}_{i,p,c,t}:$ fluxo de produto i da planta p até o cliente c no período t

$f^{WC}_{i,w,c,t}:$ fluxo de produto i do armazém w até o cliente c no período t

$f^{PP}_{i,p,q,t}:$ fluxo de produto i da planta p até a planta q no período t

$f^{PW}_{i,p,w,t}:$ fluxo do produto i planta p no warehouse w no período t

$U_{c,i,t}:$ Demanda não atendida do cliente c de produto i no período t

#Parâmetros adicionais

$S^{P}_{i,p,-1}:$ quantidade de produto i inicialmente estocado na planta p

$S^{W}_{i,w,-1}:$ quantidade de produto i inicialmente estocado no armazém w

# Parâmetros

$G_w$ : capacidade máxima de estoque do armazém w

$G_p$ : capacidade máxima de estoque da planta p

$h$ : custo de estoque em unidade/período

$k_{e,l_p,t}:$ taxa de produção max de produtos de tamanho e na linha l (da planta p) no período t

$t^{PP}_{p,q}:$ custo unitário de transporte da planta p até a planta q

$t^{PC}_{p,c}:$ custo unitário de transporte da planta p até o cliente c

$t^{WC}_{w,c}:$ custo unitário de transporte do armazém w até o cliente c

$D_{c,i,t}:$ demanda do cliente c do produto i no período t

falta taxa de transporte de planta para warehouse $t^{PW}_{p,w}$

#Função Objetivo:

$$
\text{Minimize: } \quad
\sum_{t,i} \left( \sum_{p,c} t^{PC}_{p,c} \cdot f^{PC}_{i,p,c,t} + \sum_{w,c} t^{WC}_{w,c} \cdot f^{WC}_{i,w,c,t} + \sum_{p,q} t^{PP}_{p,q}\cdot f^{PP}_{i,p,q,t}  + \sum_{w} h\cdot S^W_{i,w,t}  \right) + M\sum_{c,i,t}U_{c,i,t}
$$

falta adicionar a taxa de planta pra armazem e $t^{PW}_{p,w}f^{PW}_{i,p,w,t}$

#Restrições:

$$
\text{subject to:} \\
1) \quad \sum_{i \in e[i]} x_{i_e,l,p,t} \le k_{e,l,p,t} \quad
∀ \ p,l_p,t, e $$



$$
2) \quad S^P_{i,p,t-1} + \sum_{l}x_{i,l,p,t} + \sum_{q \neq p} f^{PP}_{i,q,p,t}  = S^P_{i,p,t} + \sum_{w}f^{PW}_{i,p,w,t} + \sum_{c}f^{PC}_{i,p,c,t} + \sum_{q \neq p} f^{PP}_{i,p,q,t} \quad \forall \ p,i,t
$$

$$
3) \quad S^W_{i,w,t-1} + \sum_{p}f^{PW}_{i,p,w,t} = S^W_{i,w,t} + \sum_{c} f^{WC}_{i,w,c,t} \quad ∀ \ p,i,t
$$

$$
4) \quad \sum_{i}S^P_{i,p,t} \le G_p \quad \forall p,t
$$

$$
5) \quad \sum_{i}S^W_{i,w,t} \le G_w \quad \forall w,t
$$

$$ 6) \quad \sum_{p}\ f^{PC}_{i,p,c,t} \ + \ \sum_{w}f^{WC}_{i,w,c,t}  + \  U_{c,i,t} = D_{c,i,t} \quad ∀ \ c,i,t $$
